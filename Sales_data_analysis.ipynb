```python
from datetime import datetime
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
file = pd.read_csv('supermarket_sales - Sheet1.csv')
```

## Data Clean Up

```python
file = file[file['Total'] >= 0]
null_value_checker = file['Total'].isnull().sum()
print(f'Number of cells where total is zero = {null_value_checker}')
```

#### Sales Data Sample 7/1000 rows displayed

```python
sum_total = file['Total'].sum()
sum_total_rounded = round(sum_total,2)
print(f'Total Sales = ${sum_total_rounded}')
```

```python
#Average of all the sales
rev_avg = file['gross income'].mean()
rev_avg_rounded = round(rev_avg,2)
print(f'Average Profit = ${rev_avg_rounded} per transaction')
```

## Total Quantity Sold (by Product Line)

```python
ag_data = file.groupby('Product line')['Quantity'].sum().reset_index()
x_val = ag_data['Product line']
y_val = ag_data['Quantity']
plt.ylabel('Quantity')
plt.xlabel('Product')
plt.xticks(rotation=45)
plt.bar(x_val,y_val)
plt.show()
```

```python
print(ag_data)
```

## Total Income (by branch)

```python
branch_rev = file.groupby('City')['gross income'].sum().reset_index()
branches_x = branch_rev['City']
branch_y = branch_rev['gross income']
plt.ylabel('Gross Income ($)')
plt.xlabel('\n City')
print(branch_rev)
plt.bar(branches_x, branch_y)
plt.show()
```

## Trends in Sales of Store A

```python
plt.style.use('seaborn-v0_8-darkgrid')
storeA = pd.read_csv('StoreA.csv')
storeA['Date'] = pd.to_datetime(storeA['Date'])
storeA.sort_values('Date', inplace=True)
ag_income = storeA.groupby("Date")['gross_income'].sum()
plt.plot(ag_income)
plt.ylabel('Gross Income ($)')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.show()
```

## Monthly Sales

```python
storeA['Date'] = pd.to_datetime(storeA['Date'])
storeA.sort_values('Date', inplace=True)
ag_income = storeA.groupby("Date")['gross_income'].sum()

'''monthly sales'''
jan_sales = ag_income.loc['2019-01-01':'2019-01-31']
feb_sales = ag_income.loc['2019-02-01':'2019-02-28']
mar_sales = ag_income.loc['2019-03-01':'2019-03-31']

jan_income = jan_sales.sum()
feb_income = feb_sales.sum()
mar_income = mar_sales.sum()

months = ['January', 'February', 'March']
average_income = [jan_income, feb_income, mar_income]


plt.bar(months, average_income)
plt.xlabel('Month')
plt.ylabel('Average Daily Income ($)')
plt.title('Average Daily Income for January, February, and March')

# Display the plot
plt.show()
print(f'January income = ${jan_income}\n'
      f'Feb income = ${feb_income}\n'
      f'March Income = ${mar_income}')
```

## Average Daily Sales by month

```python
jan_daily_income = jan_income/31
feb_daily_income = feb_income/28
mar_daily_income = mar_income/31

days_jan = list(range(1, 32))
days_feb = list(range(1, 29))
days_mar = list(range(1, 32))

months = ['January', 'February', 'March']
average_daily_income = [jan_daily_income, feb_daily_income, mar_daily_income]


plt.bar(months, average_daily_income)
plt.xlabel('Month')
plt.ylabel('Average Daily Income ($)')
plt.title('Average Daily Income for January, February, and March')

# Display the plot
plt.show()


print(f'January Per Day Average= ${jan_daily_income} \n'
      f'Feb Per Day Average= ${feb_daily_income} \n'
      f'March Per Day Average= ${mar_daily_income}')
```
